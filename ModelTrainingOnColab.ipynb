## Interpolating data as timestamps are not the same across features

```python
# Convert 'startTimeNanos' to datetime and select 'value' column for each dataframe
distance['timestamp'] = pd.to_datetime(distance['startTimeNanos'], unit='ns')
distance.set_index('timestamp', inplace=True)
distance_value = distance[['value']].rename(columns={'value': 'distance'})

step_count['timestamp'] = pd.to_datetime(step_count['startTimeNanos'], unit='ns')
step_count.set_index('timestamp', inplace=True)
step_count_value = step_count[['value']].rename(columns={'value': 'step_count'})

speed['timestamp'] = pd.to_datetime(speed['startTimeNanos'], unit='ns')
speed.set_index('timestamp', inplace=True)
speed_value = speed[['value']].rename(columns={'value': 'speed'})


combined = pd.concat([distance_value, step_count_value, speed_value], axis=1)
interpolated = combined.interpolate()
interpolated = interpolated.dropna()
```

```python
interpolated.head()
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
timestamp = interpolated.index
plt.plot(timestamp, interpolated['speed']*100, label='Speed')
plt.plot(timestamp, interpolated['distance'], label='Distance')
plt.plot(timestamp, interpolated['step_count'], label='Step Count')
plt.xlabel('Time')
plt.ylabel('Interpolated Value')
plt.title('Interpolated Value over Time')
plt.legend()
plt.show()
```

```python
df_cleaned = interpolated
```

## Building Anomaly Detection Model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# Assuming df_cleaned is your DataFrame with non-NaN rows

# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cleaned)

# Split data into train and test sets
X_train, X_test = train_test_split(scaled_data, test_size=0.2, random_state=42)

# Define the Autoencoder model
input_dim = X_train.shape[1]
encoding_dim = 8  # You can adjust the size of the encoding layer
input_layer = Input(shape=(input_dim,))
encoder_layer = Dense(encoding_dim, activation='relu')(input_layer)
decoder_layer = Dense(input_dim, activation='sigmoid')(encoder_layer)
autoencoder = Model(input_layer, decoder_layer)

# Compile the model
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Train the Autoencoder
autoencoder.fit(X_train, X_train, epochs=1000, batch_size=32, shuffle=True, validation_data=(X_test, X_test))
```

```python
import numpy as np
# Use the trained Autoencoder to reconstruct the input data
reconstructed_data = autoencoder.predict(scaled_data)

# Calculate reconstruction error
mse = np.mean(np.power(scaled_data - reconstructed_data, 2), axis=1)

# Set a threshold for anomaly detection
threshold = np.mean(mse) + 2 * np.std(mse)

# Identify anomalies
anomalies = df_cleaned[mse > threshold]
```

### Vizualizing Anomalies

```python
import matplotlib.pyplot as plt

# Assuming df_cleaned is your DataFrame with non-anomalies
# Assuming anomalies DataFrame contains the anomalies identified by the VAE model

# Plot original data points
plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned.index, scaler.inverse_transform(df_cleaned)[:, 0], color='b', label='Normal Data')

# Plot anomalies
plt.scatter(anomalies.index, scaler.inverse_transform(anomalies)[:, 0], color='r', label='Anomalies')

plt.title('Original Data with Anomalies Marked')
plt.xlabel('Time')
plt.ylabel(f'{anomalies.columns[0]}')
plt.legend()
plt.grid(True)
plt.show()
```
